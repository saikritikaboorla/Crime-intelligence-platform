##Second largest element in a list

```python
def second_largest(nums):
  unique_nums = set(nums)
  if len(unique_nums) < 2: return None
  unique_nums.remove(max(unique_nums))
  return max(unique_nums)
print(second_largest([12,45,23,45,67,12]))
```

##Frequency analysis without using count() method

```python
def frequency(num):
  unique_nums = set(num)
  print(num)
  for i in unique_nums:
    count = 0
    for j in num:
      if(i == j):
        count += 1
    print(f"frequency of {i} is {count}")
frequency([45,60,34,40,50])
```

##List Rotation

```python
def rotate(lst):
  k = 2
  return lst[-k:] + lst[:-k]
rotate([10,20,30,40,50])
```

##Remove Duplicates

```python
def remove_duplicates(nums):
  return list(dict.fromkeys(nums))
print(remove_duplicates([12, 45, 34, 12, 56]))
```

##Pair Sum

```python
a = [2, 4, 3, 5, 7, 8]
target = 10

for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == target:
            print(a[i], a[j])
```

## Longest Consecutive Sequence


```python
a = [100, 4, 200, 1, 3, 2]

s = set(a)
longest = 0

for i in s:
    if i - 1 not in s:
        count = 1
        while i + count in s:
            count += 1
        if count > longest:
            longest = count

print(longest)
```

## Matrix Operations using Nested Lists

```python
# Matrix Addition, Subtraction and Multiplication

A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

# Addition
print("Addition:")
for i in range(2):
    print([A[i][j] + B[i][j] for j in range(2)])

# Subtraction
print("Subtraction:")
for i in range(2):
    print([A[i][j] - B[i][j] for j in range(2)])

# Multiplication
print("Multiplication:")
C = [[0, 0], [0, 0]]

for i in range(2):
    for j in range(2):
        for k in range(2):
            C[i][j] += A[i][k] * B[k][j]

for row in C:
    print(row)
```

## Spiral Traversal

```python
a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

top, bottom = 0, len(a)-1
left, right = 0, len(a[0])-1

while top <= bottom and left <= right:
    for i in range(left, right+1):
        print(a[top][i], end=" ")
    top += 1

    for i in range(top, bottom+1):
        print(a[i][right], end=" ")
    right -= 1

    for i in range(right, left-1, -1):
        print(a[bottom][i], end=" ")
    bottom -= 1

    for i in range(bottom, top-1, -1):
        print(a[i][left], end=" ")
    left += 1
```

##  Majority Element

```python
a = [2, 2, 1, 2, 3, 2, 2]

found = False

for i in a:
    if a.count(i) > len(a) // 2:
        print(i)
        found = True
        break

if not found:
    print("No Majority Element")
```

## Stock Buy and Sell

```python
prices = [7, 1, 5, 3, 6, 4]

buy = prices[0]
profit = 0

for i in prices:
    if i < buy:
        buy = i
    elif i - buy > profit:
        profit = i - buy

print(profit)
```
